```python
import pandas as pd
import string
from matplotlib import pyplot as plt
%matplotlib inline
stats = pd.read_csv("statistics.csv")
```

```python
stats[stats["Project"] == "<all>"]
```

```python
df = pd.read_excel("../Dataset-SBES2020.xlsx")
df = df[(df["Committers B1"] != 0) & (df["Committers B2"] != 0)]

df.loc[df["Conflicts"] != "YES", 'Conflicts'] = "WO"
df.loc[df["Conflicts"] == "YES", 'Conflicts'] = "WC"
YES = "YES"
```

```python
df
```

```python
attributes = {
    "Branching-duration": "branching-duration",
    "Total-duration": "total-duration",
    "Commits B1": "commits B1",
    "Commits B2": "commits B2",
    "Committers B1": "committers B1",
    "Committers B2": "committers B2",
    "Changed Files B1": "changed-files B1",
    "Changed Files B2": "changed-files B2"
}

for i, attr in enumerate(attributes):
    letter = string.ascii_lowercase[i]
    nattr = attributes[attr].replace('-', '_').replace(' ', '_')
    fig, ax2 = plt.subplots()
    p = df.boxplot(ax=ax2, grid=False, vert=False, widths = 0.5, showfliers=False, column=attr, by="Conflicts")
    fig.suptitle('')
    ax2.set_title("")
    xlim = ax2.get_xlim()
    ax2.set_xlabel(f"")
    fig.set_size_inches([3, 1])
    plt.tight_layout()
    print(nattr)

    display(fig)
    fig.savefig(f"output/distributions/{letter}_{nattr}_dist.png")
    fig.savefig(f"output/distributions/{letter}_{nattr}_dist.svg")
    fig.savefig(f"output/distributions/{letter}_{nattr}_dist.pdf")
    plt.close()
```
